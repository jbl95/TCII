<b>3)</b> Simulación matemática utilizando valores normalizados.

```python
from splane import pzmap, grpDelay, bodePlot
from scipy.signal import TransferFunction,bode
import matplotlib.pyplot as plt
import numpy as npy

## Simulación matemática.

wo = 1
R = 1
C = npy.sqrt(2) / 2
L = npy.sqrt(2)

K = 1 / (L*C)
a = 1 / (C*R)

num = [K]
den = [1,a,K]
 
my_tf = TransferFunction( num, den )

## Ploteo.
bodePlot(my_tf)
pzmap(my_tf)
```

<b>Comentario:</b> Para este caso, se presentaba una topología circuital que debía cumplir con las especificaciones de máxima planicidad para un $\epsilon = 1$ (caso Butterworth). Según lo simulado en LTspice se observa una caida de 3dB en 1KHz que para nuestra simulación matemática corresponde con $w_o = 1$.
