```python
from splane import pzmap, grpDelay, bodePlot
from scipy.signal import TransferFunction,bode
import matplotlib.pyplot as plt
import numpy as npy
```

```python
## Simulación matemática.

R3 = 1000
C = 1e-6

p = 1 / (R3 * C)

num = [1,-p]
den = [1,p] 

my_tf = TransferFunction( num, den )

## Ploteo.
bodePlot(my_tf)
pzmap(my_tf)
```

```python
## Simulación matemática - Normalizada.
p_norm = 1
num_n = [1,-p_norm]
den_n = [1,p_norm] 

my_tf = TransferFunction( num_n,den_n)

## Ploteo.
bodePlot(my_tf)
pzmap(my_tf)
```

Para la normalización calculada de Wa = 1 / C*R3 se obtiene una respuesta en frecuencia similar.</div>

```python
from splane import pzmap, grpDelay, bodePlot
from scipy.signal import TransferFunction,bode
import matplotlib.pyplot as plt
import numpy as npy
```

```python
## Simulación matemática.

R = 1000
RA = 1000
RB = 5000
C = 1e-6

K = RA / RB

a = ((2*K-1) / K) * (1/C*R)
b = (1/C*R) ** 2
c = (3/C*R)
d = b

num = [-K,-K*a,-K*b]
den = [1,c,d]
 
my_tf = TransferFunction( num, den )

## Ploteo.
bodePlot(my_tf)
pzmap(my_tf)
```

```python
## Simulación matemática - Normalizada.

num_n = [1,-3,1]
den_n = [1,3,1] 

my_tf = TransferFunction( num_n,den_n)

## Ploteo.
bodePlot(my_tf)
pzmap(my_tf)
```
