```python
from splane import pzmap, grpDelay, bodePlot
from scipy.signal import TransferFunction,bode
import matplotlib.pyplot as plt
import numpy as npy
```

```python
## Simulación matemática.

R3 = 1000
C = 1e-6

p = 1 / (R3 * C)

num = [1,-p]
den = [1,p] 

my_tf = TransferFunction( num, den )

bodePlot(my_tf)
pzmap(my_tf) #S plane pole/zero plot

```

```python
## Simulación matemática - Normalizada.
p_norm = 1
num_n = [1,-p_norm]
den_n = [1,p_norm] 

my_tf = TransferFunction( num_n,den_n)

bodePlot(my_tf)
pzmap(my_tf) #S plane pole/zero plot
```
