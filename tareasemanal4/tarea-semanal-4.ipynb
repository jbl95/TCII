```python
from splane import pzmap, grpDelay, bodePlot
from scipy.signal import TransferFunction,bode
import matplotlib.pyplot as plt
import numpy as npy

## Simulación matemática.
a = 1
b = npy.sqrt(2)
c = 0
d = 3
k = 1
qq = npy.sqrt(2) / 2
w_o = 1

zz = (w_o/qq) * (a - (b*k*qq))
yy = (a - ((c-d)*k))

a = [a,zz * (w_o / qq),yy] 
b = [1,w_o/qq,w_o**2]

my_tf = TransferFunction(a,b)

## Ploteo.
bodePlot(my_tf)
pzmap(my_tf)
```

```python
from splane import pzmap, grpDelay, bodePlot
from scipy.signal import TransferFunction,bode
import matplotlib.pyplot as plt
import numpy as npy

## Simulación matemática.
C1 = 1
C2 = 1
L1 = 1
L2 = 1
R1 = npy.sqrt(2)
R2 = 1/npy.sqrt(2)

K = L2 / L1

a = [K*1,K*R2/L2,K*1/(C2*L2)]
b = [1,R1/L1,1/(C1*L1)]

my_tf = TransferFunction(a,b)

## Ploteo.
bodePlot(my_tf)
pzmap(my_tf)
```
